# 1. Los top 10 tweets m谩s retweeted

Haremos la funci贸n que retorna los tweets con mayor cantidad de retweets (`retweetCount`), junto con un ejemplo con la muestra

```python
def topRetweets(df):
    return df.nlargest(10, 'retweetCount')[['content', 'retweetCount','user']]
```

```python
topRetweets(sample)
```

# 2. Los top 10 usuarios en funci贸n a la cantidad de tweets que emitieron.

Haremos una funci贸n que retorne los 10 usuarios con m谩s cantidad de tweets, contando la aparici贸n de `user` en el `dataframe`

fuente: https://datascientyst.com/get-most-frequent-values-pandas-dataframe/

```python
def topUsers(df):
    users = df['user'].value_counts().index.tolist()[:10]
    return list(map(lambda x: x['username'], users))
```

```python
topUsers(sample)
```
